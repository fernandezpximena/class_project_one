```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd

Suc = "Sucursales.csv"
Sucursales = pd.read_csv(Suc)
Rez = "RezagoFinal.csv"
Rezago = pd.read_csv(Rez)
pd.set_option('display.max_columns', None)
Rezago[["Clave Municipio","Clave Estado","Población 2015"]]=Rezago[["Clave Municipio","Clave Estado","Población 2015"]].astype(str)
Rezago['Clave Estado'] = Rezago['Clave Estado'].apply(lambda x: x.split('.')[0])
Rezago['Clave Municipio'] = Rezago['Clave Municipio'].apply(lambda x: x.split('.')[0])
Rezago['Población 2015'] = Rezago['Población 2015'].apply(lambda x: x.split('.')[0])
```

```python
display(Sucursales.head (2))
display(Rezago.head (2))
```

```python
SR=pd.merge(left=Rezago, right=Sucursales, on="Clave Estado-Municipio", how="outer" )
SRdrop = SR.dropna(how='any')
#SRdrop.isnull().sum()
SR.head(2)
```

```python
SRdrop.isnull().sum()
```

```python
SRx = SRdrop.drop(SRdrop.columns[[12, 13, 14, 15]], axis=1)
SRFinal=SRx.rename(columns = {"Clave Estado_x":"Clave Estado", "Clave Municipio_x":"Clave Municipio", "Estado_x":"Estado","Municipio_x":"Municipio"})
SRFinal.sort_values(by=["Estado"], ascending=True)
SRFinal.head(2)
```

```python
SRFinal.to_csv("Sucursales y Rezago.csv", index=False, header=True)
```

```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd

SR = "Sucursales y Rezago.csv"
pd.set_option('display.max_columns', None)
df = pd.read_csv(SR)
df.head(1)
```

```python
maker_group = df.groupby('Estado')
count_makers = maker_group['Tipo de Servicio'].count()
count_chart = count_makers.plot(kind='bar')
count_chart.set_xlabel("Estado")
count_chart.set_ylabel("Servicios financieros disponibles")
plt.show()
plt.tight_layout()
```

```python
mg2 = df.groupby('Tipo de Servicio')
cm2 = mg2['Estado'].count()
cc2 = cm2.plot(kind='pie')
cc2.set_xlabel("Estado")
cc2.set_ylabel("Servicios financieros disponibles")
plt.show()
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,8))
plt.pie(df['Tipo de Servicio'].value_counts(),labels=df['Tipo de Servicio'].unique(),autopct='%1.1f%%')
```

```python
fig = plt.figure(figsize=(7,7))
plt.pie(df['Grado de Rezago Social 2015'].value_counts(),labels=df['Grado de Rezago Social 2015'].unique(),autopct='%1.1f%%')
```
